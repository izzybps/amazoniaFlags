```python
import requests
import json
import shutil
import os
```

```python
amazonStates = [
    {'value':100112, 'data-estado':"AC"},
    {'value':100113, 'data-estado':"AM"},
    {'value':100116, 'data-estado':"AP"},
    {'value':100121, 'data-estado':"MA"},
    {'value':100151, 'data-estado':"MT"},
    {'value':100115, 'data-estado':"PA"},
    {'value':100111, 'data-estado':"RO"},
    {'value':100114, 'data-estado':"RR"},
    {'value':100117, 'data-estado':"TO"}]
```

```python
headers = {
'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
}

for state in amazonStates:
    state_value = str(state["value"])
    get_city_url = f'https://www.cnm.org.br/institucional/get_municipios/{state_value}'

    # Check if the directory exists
    path_state = os.path.abspath(f'./{state["data-estado"]}/')
    if not os.path.exists(path_state):
        # If it doesn't exist, create it
        os.makedirs(path_state)

    r = requests.get(get_city_url, headers=headers)
    for city in r.json():
        get_flag_url = f'https://www.cnm.org.br/public/assets/img/bandeiras/{state_value}/{city["id"]}/1.png'
        get_flag = requests.get(get_flag_url, stream=True)
        
        cod_ibge = '0'

        r = requests.get(f'https://www.cnm.org.br/municipios/get_results/{state_value}/{city["id"]}', headers=headers)
        if 'codIbge' in r.text:
            cod_ibge = r.text[r.text.find('codIbge')+8: r.text.find('codIbge')+20].split('"')[0]
        else:
            print('não foi encontrado codigo ibge na página')

        img_path = f'./{state["data-estado"]}/{cod_ibge}.png'
        
        if get_flag.status_code == 200:
            with open(img_path, 'wb') as f:
                get_flag.raw.decode_content = True
                shutil.copyfileobj(get_flag.raw, f)
                print(f'Flag from {city["nome"], cod_ibge} downloaded')
        else:
            print("failed to request flag")
        
```
